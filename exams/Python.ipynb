```python
# Write a function count_vowels(s) that
# returns the number of vowels in the string s.
```

```python
def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for char in s:
        if char in vowels:
            count+=1
    return count
```

```python
count_vowels("hello")
```

```python
# Write a function tuples_to_dict(tuples)
# that returns a dictionary with first tuple
# elements as keys and second tuple
# elements as values. For example
# tuples_to_dict([('a', 1), ('b', 2)]) returns {'a':
# 1, 'b': 2}.
```

```python
def tuples_to_dict (tuples):
    result = {}
    for key, values in tuples:
        result[key]=values
    return result
```

```python
print(tuples_to_dict([('a', 1), ('b', 2)]))
```

```python
# Write a function
# <code>count_tuples (tuples)</code> that
# returns a dictionary with tuple elements as
# keys and their counts as values.
```

```python
def count_tuples(tuples):
    result = {}
    for key in tuples:
        if key in result:
            result[key] += 1
        else:
            result[key] = 1
    return result
```

```python
data = [('a', 1), ('b', 2), ('a', 1), ('c', 3), ('b', 2), ('a', 1)]
print(count_tuples(data))
```

```python
# Write a function merge_dicts(d1, d2)
# that merges two dictionaries of lists,
# appending values for matching keys.
```

```python
def merge_dicts(d1,d2):
    for key in d2:
        if key in d1:
            d1[key]=d1[key]+d2[key]
        else:
            d1[key]=d2[key]
    return d1
```

```python
d1 = {'a': [1, 2], 'b': [3,4]}
d2 = {'b': [4, 5], 'c': [6]}

print(merge_dicts(d1, d2))
```

```python
# Write a function group_words(words)
# that returns a dictionary where keys are
# first letters and values are lists of words
# starting with that letter.
```

```python
def group_words(words):
    result = {}
    for word in words:
        first = word[0]
        if first in result:
            result[first]+=[word]
        else:
            result[first]= [word]
    return result
    
```

```python
print(group_words(["apple", "banana", "avocado", "blueberry", "cherry"]))
```

```python
# Write a function count_case(s) that
# returns a tuple with the number of
# uppercase and lowercase letters in the
# string s.
```

```python
def count_case(s):
    upper = 0
    lower = 0
    for char in s:
        if 'A' <= char <= 'Z':
            upper+=1
        elif 'a' <= char <= 'z':
            lower+=1
    return (upper,lower)
```

```python
print(count_case("Hello World!"))
```

```python
def common_elements(a, b):
    result = []
    for elem in a:
        if elem in b and elem not in result:   # common and not already added
            result = result + [elem]
    return result
```

```python
# Write a function
# has_consecutive_triplet(lst) that
# returns True if any three consecutive
# numbers in the list are strictly
# increasing or decreasing.
```

```python
def has_consecutive_triplet(lst):
    # stop at len(lst)-2 so we can safely access i, i+1, i+2
    for i in range(len(lst) - 2):
        a, b, c = lst[i], lst[i+1], lst[i+2]
        if a < b < c or a > b > c:   # strictly increasing or decreasing
            return True
    return False
```

```python
print(has_consecutive_triplet([1, 2, 3, 2, 1]))   # True (1<2<3)
print(has_consecutive_triplet([5, 4, 3, 4]))      # True (5>4>3)
print(has_consecutive_triplet([1, 1, 1, 2]))      # False (not strictly inc/dec)
```

```python
# Write a function flatten(nested_list)
# that takes a list of lists and returns a
# single flattened list.
```

```python
def flatten(nested_list):
    flat = []
    for sublist in nested_list:
        for item in sublist:
            flat = flat + [item]
    return flat
```

```python
print(flatten([[1, 2], [3, 4], [5]]))
```

```python
# Write a function sum_dict_values(d) that
# returns a new dictionary where each key's
# value is the sum of its list elements.
```

```python
def sum_dict_values(d):
    result = {}
    for key in d:
        total = 0
        for num in d[key]:
            total+=num
        result[key]=total
    return result
```

```python
data = {'a': [1, 2, 3], 'b': [10, 20], 'c': []}
print(sum_dict_values(data))
```

```python
# A library book management system is organized as a dictionary where the keys are book titles and
# the values are lists containing the author and the number of available copies.
# Implement the Python function query(library, author, number) that returns:
# All the books by the specified author
# All books with fewer than a certain number of copies available
```

```python
def query(library,author,number):
    by_author = []
    low_stock = []
    for title, info in library.items():
        book_author,copies = info[0],info[1]
        if book_author == author:
            by_author+=[title]
        if copies < number:
            low_stock = low_stock + [title]
    return by_author, low_stock
```

```python
# Sample library dictionary
library = {
    "1984": ["George Orwell", 5],
    "To Kill a Mockingbird": ["Harper Lee", 2],
    "The Great Gatsby": ["F. Scott Fitzgerald", 3]
}

# Example usage
result = query(library, "George Orwell", 4)
print(result)
```

```python
# Write a function remove_duplicates(lst) that removes duplicates while preserving the first occurrence of each element.
```

```python
def remove_duplicates(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result +=[item]
    return result
```

```python
# Write a function is_anagram(s1, s2) that returns True if the two strings are anagrams (contain the same characters in any order).
```

```python
def is_anagram(s1,s2):
    if len(s1)!=len(s2):
        return False

    count = {}
    for char in s1:
        count[char]=count.get(char,0)+1 
    for char in s2:
        if char not in count:
            return False
        count[char]-=1
        if count[char]<0:
            return False

    return True
```

```python
print(is_anagram("listen", "silent"))  # True
print(is_anagram("triangle", "integral"))  # True
print(is_anagram("apple", "pale"))     # False
```

```python
# Write a python code for pangram
```

```python
def is_pangram(s):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    s = s.lower()
    for letter in alphabet:
        if letter not in s:
            return False
    return True
```

```python
sentence1 = "The quick brown fox jumps over the lazy dog"
sentence2 = "Hello World"

print(is_pangram(sentence1))  # True
print(is_pangram(sentence2))  # False
```

```python
# Implement the Python function ds_oddmax(xlist) that returns a list containing the maximum elements
# contained in the inner lists having an odd number of elements.This question requires that exactly 
# only (ONLY) the function must be implemented. Therefore, no data structure nor function call are 
# requested from the candidate.
```

```python
def ds_oddmax(xlist):
    result = []
    for inner in xlist:
        n = len(inner)
        if n % 2 != 0:
            max_val = inner[0]
            for num in inner:
                if num > max_val:
                    max_val = num
            # Concatenate to result
            result = result + [max_val]
    return result
```

```python
xlist = [[1, 7, 12], [4, 2, 10, 14], [15, 3, 2, 11, 20]]
print(ds_oddmax(xlist))
```

```python
# Write a program that asks the user for a string and prints it all in uppercase letters
```

```python
s = input("Type a string: ")
uppercase = ""
for this in s:
    if 'a' <= this <= 'z':
        uppercase_char = chr(ord(this)-32)
    else:
        uppercase_char = this
    uppercase += uppercase_char
print("Uppercase string:", uppercase)
```

```python
#Define a function apply_discount(price,discount_percent) that takes a price and a discount percent, and returns the discounted price rounded to 2 
# decimal points
```

```python
def apply_discount(price, discount_percent):
    discounted_price = price * (1- discount_percent/100)
    discounted_price = int(discounted_price*100+0.5)/100
    return discounted_price
    
```

```python
print(apply_discount(100,10.2))
```
