```python
# Input the text and the word
text = input("Enter the text: ")
word = input("Enter the word to find: ")

# Function to calculate the percentage of a word in the text
def percent_word(text, word):
    text = text.lower().split()  # Convert text to lowercase and split into words
    word = word.lower()          # Ensure the word is also lowercase
    total_words = len(text)      # Total number of words in the text
    count = text.count(word)     # Count occurrences of the word
    if total_words == 0:         # Prevent division by zero
        return 0
    percentage = (count / total_words) * 100  # Calculate the percentage
    return round(percentage, 2)  # Round to 2 decimal places

# Output the result
result = percent_word(text, word)
print(f"The word '{word}' makes up {result}% of the text.")
```

```python
'''Linked List
You are making a Music Player, which allows you to create a playlist of tracks.
The given code defines Player and Track classes, where the Player is a linked list, chaining together Track objects.
The code takes a number of tracks from user input and adds them to the playlist.
You need to iterate over the linked list and output all the tracks in the playlist in the order of playback.'''
class Track:
    def __init__(self, title, next):
        self.title = title  # The name of the track
        self.next = next    # Points to the next track in the playlist

class Player:
    def __init__(self):
        self.head = None    # Initially, the playlist is empty (no tracks)

    def add(self, title):
        if not self.head:     # If the playlist is empty
            self.head = Track(title, None)   # Create the first track
            return
        curr = self.head
        while curr.next:    # Traverse the linked list until the last track
            curr = curr.next
        curr.next = Track(title, None)    # Add the new track at the end


p = Player()    # Create a new Player object
while True:
    x = input("Enter a track (or 'end' to stop): ")
    if x == 'end':     # Stop taking input when the user enters 'end'
        break
    p.add(x)   # Add the entered track to the playlist


curr = p.head     # Start at the first track
while curr:
    print(curr.title)     # Print the title of the current track
    curr = curr.next      # Move to the next track
```

```python
def create():
    llist=dict()                    # Create an empty dictionary to represent the linked list
    llist['head']='tail'            # Initialize the head pointer pointing to 'tail'
    llist['tail']=0                 # Represent the end of the list with 'tail'
    return llist
def add_node(llist,newNode):
    llist[newNode]=llist['head']    # Set the new node's pointer to the current head
    llist['head']=newNode           # Update head to point to the new node
    return llist
def ltraverse(llist):
    current='head'                  # Start traversal at the head
    path=[]
    path.append((0,current))        # Add the head node
    k=1
    while current != 'tail':        # Keep traversing until reaching the tail
        next=llist[current]         # Get the next node
        path.append((k,next))       # Add the current node to the path
        current=next                # Move to the next node
        k+=1
    return path
def delete_first(llist):
    current=llist['head']           # Get the current head node
    newFirst=llist[current]         # Get the next node (new head)
    llist['head']=newFirst          # Update head to point to the new first node
    del llist[current]              # Remove the old head node
    return llist
def find(llist,node):
    current='head'                  # Start at the head
    while current != 'tail':        # Traverse until reaching the tail
        next=llist[current]         # Get the next node
        if next == node:            # If the node is found, return True
            return True
        current=next                # Move to the next node
    return False                    # Node not found
def count_llist(llist):
    current='head'                  # Start at the head
    total=0
    while current != 'tail':        # Traverse until reaching the tail
        next=llist[current]
        current=next                # Move to the next node
        total+=1                    # Increment the counter
    return total
l=create()
listt=[10, 20, 30, 40]
for x in reversed(range(len(listt))):# Add nodes in reverse order
    add_node(l,listt[x])

print(find(l,10))                    # True (10 is in the linked list)
print(l)                             # {'head': 40, 'tail': 0, 10: 'tail', 20: 10, 30: 20, 40: 30}
print(count_llist(l))                # 4 (Nodes: 40 -> 30 -> 20 -> 10)
delete_first(l)                      
print(l)                             # {'head': 30, 'tail': 0, 10: 'tail', 20: 10, 30: 20}
```

```python
'''A graph is stored on an edgelist implemented as a list of tuples. Each tuple contains two values, representing the nodes at the ends of an edge. 
Implement the Python function neighbors2(graph) that returns the node(s) having exactly two connections with other nodes. For example:
graph = [(1, 4), (1, 3), (2, 3), (2, 5)].'''

def neighbors2(graph):
    degsum = dict()        # Dictionary to store the number of connections (degree) of each node           
    for u,v in graph:      # Loop through each edge (u, v)
        degsum[u]=degsum.get(u,0)+1 # Increment count for node u
        degsum[v]=degsum.get(v,0)+1 # Increment count for node v
    result = []                     # List to store nodes with exactly 2 connections
    for k in degsum:                # Loop through all nodes in the dictionary
        if degsum[k]==2:            # Check if the node has exactly 2 connections
            result.append(k)        # Add the node to the result list
    return result                   # Return the final list of nodes
graph = [(1, 4), (1, 3), (2, 3), (2, 3),(2,4),(6,7)]

# Call the function
result = neighbors2(graph)

# Print the result
print(result)


```

```python
''' 
btree = { 'root':{'P':None, 'L': 1, 'R':10},
        1:{'P':'root', 'L':5, 'R':3},
        10:{ 'P':'root', 'L':12, 'R':13},
        5:{'P':1, 'L':None, 'R':None},
        3:{'P':1, 'L':None, 'R':None},
        12:{'P':10, 'L':None, 'R':None),
        13:{'P':10, 'L':None, 'R':None}}
A binary tree is implemented using a dictionary of dictionaries such as in the example above. Implement the Python function higher(tree, root, high)
that returns the nodes having values higher than high. The structure must be traversed using the references. '''

def higher(tree, root, high):
    queue = [root]
    visited = []
    while queue:
        current = queue.pop(0)
        lc = tree [current]['L']
        rc = tree [current]['R']
        if lc != None:
            queue.append(lc)
        if rc!= None:
            queue.append(rc)
        if isinstance(current,(int,float)):
            if current > high:
                visited.append(current)
    return visited
```

```python
''' 
A stack is stored in a doubly linked list implemented using a dictionary of dictionaries like, for example,
dll = { 'head': {'prev': None, 'next': 12},
12: {'prev': 'head', 'next': 20},
20: {'prev': 12, 'next': 15},
15: {'prev': 20, 'next':2},
2: {'prev': 15, 'next': 'tail'},
'tail': {'prev': 2, 'next': None}}
Implement the Python function stack3(listx, v1) that: 1) creates an empty doubly linked list, 2) inserts the (possibly many) 
elements contained in the function argument so as to implement a stack, 3) inserts the argument v1 into the stack and removes
two elements from the stack, 4) returns the doubly linked list. '''

def stack3(listx,v1):
    dll = {'head':{'prev':None,'next':'tail'},
          'tail':{'prev':'head','next':None}}
    listx.append(v1)
    for x in listx:
        succ=dll['head']['next']
        dll[x]={'prev':'head','next':succ}
        dll['head']['next']= x
        dll[succ]['prev']=x
    for x in range(2):
        del1 = dll['head']['next']
        ln = dll[del1]['prev']
        rn = dll[del1]['next']
        dll[ln]['next']=rn
        dll[rn]['prev']=ln
        del dll[del1]
    return dll
```

```python
#Implement a function that returns the sum of all the maxes in the lists
a = [[1,2,4],[2,8,9],[3,5,6]]
def summaxes(a):
    total_sum = 0  # Variable to store the total sum of maxes
    for x in a:  # Loop through each sublist
        max_value = x[0]  # Assume the first element is the maximum
        for i in x:  # Loop through the current sublist
            if i > max_value:  # Update `max_value` if a larger value is found
                max_value = i
        total_sum += max_value  # Add the max value of this sublist to the total sum
    return total_sum  # Return the total sum
print(summaxes(a))
```

```python
''' Iterative programming: Given the python list array composed with n integer numbers (not distinct).
Implement the function frequency(array) that returns how many times each element of array occurs. '''
def frequency(array):
    l = len(array)                 #create the variable l that stores the length of the array
    mp = {}                        #creates the dictionary (which will store the result)
    for x in range (l):
        if array[x] not in mp:     #checks if each element is in the dictionary
            mp[array[x]]=0         #if not add and assign a value of 0
            print(mp)
        mp[array[x]]+=1            #increment the number that is assigned (occurence)
    for i in mp:                   #for printing
        print('Element',i,'occurs',mp[i],'times')  #i is the key,mp[i] is the value(number of occurences)
    return
print(frequency([12,3,45,12,12,34,45]))
```

```python
''' 
A doubly linked list is implemented using a dictionary of dictionaries such as, for
example:
listx = { 'head': {'prev': None, 'next': 12}, 
12:{'prev': 'head', 'next': 20}, 
20: {'prev': 12,'next': 15},
15: {'prev': 20, 'next': 2},
2 : {'prev': 15, 'next': 'tail'},
'tail': {'prev': 2, 'next': None}}
with two sentinel nodes (head and tail) and a collection of nodes 12, 20, 15, 2.
Implement the Python function
poplast2 (listx) that removes the two nodes immediately preceding the tail and returns the doubly linked list after removals. The
structure must be traversed using the references.  '''
def poplast2(listx):
    # Find the node immediately before the tail
    node_before_tail = listx['tail']['prev']
    
    # If there's no node or only one node before the tail, adjust accordingly
    if node_before_tail is None:
        return listx  # Nothing to remove
    
    # Find the second node before the tail
    second_node_before_tail = listx[node_before_tail]['prev']
    
    if second_node_before_tail is None:
        # If there's only one node before the tail, remove it
        listx['head']['next'] = 'tail'
        listx['tail']['prev'] = 'head'
        del listx[node_before_tail]
        return listx
    
    # Adjust the links to skip the last two nodes before the tail
    listx[second_node_before_tail]['next'] = 'tail'
    listx['tail']['prev'] = second_node_before_tail
    
    # Delete the two nodes
    del listx[node_before_tail]
    del listx[listx['tail']['prev']]  # Use the updated `prev` for tail to delete the second last node
    
    return listx
#Example
listx = { 
    'head': {'prev': None, 'next': 12}, 
    12: {'prev': 'head', 'next': 20}, 
    20: {'prev': 12, 'next': 15},
    15: {'prev': 20, 'next': 2}, 
    2: {'prev': 15, 'next': 'tail'},
    'tail': {'prev': 2, 'next': None}
}

updated_list = poplast2(listx)
print(updated_list)
```

```python
'''  
A doubly linked list is implemented using a dictionary of dictionaries such as, 
for example: 
listx = { 'head': {'prev': None, 'next: 12),
12: ('prev': 'head', 'next': 20),
20: {'prev: 12, 'next: 15), 
15: {'prev: 20, 'next: 2),
2: ('prev: 15, 'next': 'tail'),
'tail': {'prev: 2, 'next': None}}
with two sentinel nodes (head and tail) and a collection of nodes 12, 20, 15, 2.
Implement the Python function dllc(listx) that returns the sum of the numeric values stored
in the nodes of the doubly linked list. The references of the doubly linked list must be preserved.
The structure must be traversed using the references.  '''
def dllc(listx):
    current_node = listx['head']['next']  # Start from the first node after 'head'
    total_sum = 0
    
    # Traverse the list until 'tail' is reached
    while current_node != 'tail':
        total_sum += current_node  # Add the numeric value of the current node
        current_node = listx[current_node]['next']  # Move to the next node
    
    return total_sum
#example
listx = { 
    'head': {'prev': None, 'next': 12}, 
    12: {'prev': 'head', 'next': 20}, 
    20: {'prev': 12, 'next': 15},
    15: {'prev': 20, 'next': 2}, 
    2: {'prev': 15, 'next': 'tail'},
    'tail': {'prev': 2, 'next': None}
}

result = dllc(listx)
print(result)
```

```python
''' 
btree= { 'root':{'P':None, 'L': 1, 'R':10),
1:{'P': 'root', 'L':5, 'R':3), 
10:{ 'P':'root', 'L:12, 'R':13),
5:['P':1, 'L':None, 'R:None),
3:{'P':1, 'L':None, 'R:None),
12:{'P':10, 'L':None, 'R':None),
13:{'P':10, 'L':None, 'R':None}}
A binary tree is implemented using a dictionary of dictionaries such as in the example above. 
Implement the Python function btc(tree, root) that returns the sum of the values stored in the 
nodes of the binary tree. Only numeric values must be considered. The structure must be traversed
using the references.   '''
def btc(tree, root):
    if root is None:
        return 0  # Base case: no node to process
    
    # If the root is numeric, add it to the sum; otherwise, treat it as 0
    total = root if isinstance(root, (int, float)) else 0
    
    # Traverse left and right subtrees using the references
    left_child = tree[root].get('L')
    right_child = tree[root].get('R')
    
    total += btc(tree, left_child)
    total += btc(tree, right_child)
    
    return total
#example
btree = { 
    'root': {'P': None, 'L': 1, 'R': 10},
    1: {'P': 'root', 'L': 5, 'R': 3}, 
    10: {'P': 'root', 'L': 12, 'R': 13},
    5: {'P': 1, 'L': None, 'R': None}, 
    3: {'P': 1, 'L': None, 'R': None},
    12: {'P': 10, 'L': None, 'R': None}, 
    13: {'P': 10, 'L': None, 'R': None}
}

result = btc(btree, 'root')
print(result)

```
