### PATTERN 1 — Sublist row formula
From your notes: `a[x] = a[x][0] + a[x][1] - a[x][2]`  
The trick: treat each row as a mini-list and apply a formula to collapse it into a single number.

````python
# WORKED EXAMPLE: replace each sublist with sum of first two minus last element
a = [[2, 5, 1],
     [4, 3, 6],
     [1, 1, 0]]

for x in range(len(a)):
    a[x] = a[x][0] + a[x][1] - a[x][2]
print(a)  # [6, 1, 2]
````

### PATTERN 2 — Count characters / elements with a condition

````python
# WORKED EXAMPLE: count how many digits are in a string
def count_digits(text):
    count = 0
    for ch in text:
        if '0' <= ch <= '9':   # same trick used in your uppercase converter
            count += 1
    return count

print(count_digits("hello123world456"))  # 6
````

````python
# PRACTICE 2: Implement count_consonants(text) that returns the number of consonants.
# Consonants = letters that are NOT vowels (ignore spaces and punctuation).
# count_consonants("Hello World") => 7

# YOUR CODE HERE
````

````python
# SOLUTION
def count_consonants(text):
    vowels = 'aeiou'
    count = 0
    text = text.lower()
    for ch in text:
        if 'a' <= ch <= 'z' and ch not in vowels:
            count += 1
    return count

print(count_consonants("Hello World"))  # 7
````

### PATTERN 4 — summax / summins on sublists  ← exam favourite

````python
# WORKED EXAMPLE (from your notes): sum of all maxes in sublists
def summaxes(a):
    total_sum = 0
    for x in a:
        max_value = x[0]
        for i in x:
            if i > max_value:
                max_value = i
        total_sum += max_value
    return total_sum

print(summaxes([[1,2,4],[2,8,9],[3,5,6]]))  # 4+9+6 = 19
````

# DAY 2 — Dictionaries
### Core patterns:
- **Build:** `d[key] = value` or `d[key] = d.get(key, 0) + 1`
- **Iterate:** `for key in d:` or `for key, val in d.items():`
- **Check:** `if key in d:`
- **Invert:** swap keys ↔ values
- **Group:** values that share something become a list under one key

````python
# WORKED EXAMPLE: count letter frequency (your exam pattern)
def count_letters(text):
    d = dict()
    text = text.lower()
    for x in text:
        if x in d:
            d[x] += 1
        else:
            d[x] = 1
    return d

print(count_letters('banana'))
# {'b':1, 'a':3, 'n':2}
````

````python
# PRACTICE 6: Implement word_lengths(text) that returns a dictionary where each key is a word
# and the value is the length of that word.
# word_lengths("hello world foo") => {'hello':5, 'world':5, 'foo':3}

# YOUR CODE HERE
````

````python
# SOLUTION
def word_lengths(text):
    d = dict()
    words = text.split()
    for w in words:
        d[w] = len(w)
    return d

print(word_lengths("hello world foo"))  # {'hello':5, 'world':5, 'foo':3}
````

````python
# PRACTICE 8: Implement invert_and_group(d) — same as invert_dic from your notes.
# The values become keys; multiple original keys that shared the same value are grouped in a list.
# invert_and_group({'x':1,'y':2,'z':1}) => {1:['x','z'], 2:['y']}

# YOUR CODE HERE
````

````python
# SOLUTION
def invert_and_group(d):
    inverse = dict()
    for k in d:
        val = d[k]
        if val not in inverse:
            inverse[val] = [k]
        else:
            inverse[val].append(k)
    return inverse

print(invert_and_group({'x':1,'y':2,'z':1}))  # {1:['x','z'], 2:['y']}
````

````python
# WORKED EXAMPLE: set operations (your exam pattern)
eu_members = {'Germany', 'France', 'Italy', 'Spain'}
nato_members = {'Germany', 'France', 'USA', 'UK', 'Italy'}

print(eu_members & nato_members)         # in both: {'Germany','France','Italy'}
print(eu_members | nato_members)         # all combined
print(nato_members - eu_members)         # in NATO but not EU
print({'Germany','France'}.issubset(eu_members))  # True
````

````python
# WORKED EXAMPLE: insertion sort (your exam pattern — memorize this)
def insertionSort(a):
    for j in range(1, len(a)):    # start from index 1
        key = a[j]                # element to be inserted
        i = j
        while i > 0 and a[i-1] > key:  # shift larger elements right
            a[i] = a[i-1]
            i = i - 1
        a[i] = key                # place element in correct position
    return a

print(insertionSort([5, 2, 8, 1, 9]))  # [1, 2, 5, 8, 9]
````

````python
# PRACTICE 10: Write insertionSort_desc(a) that sorts in DESCENDING order.
# Change only ONE character in the while condition.
# insertionSort_desc([5,2,8,1,9]) => [9,8,5,2,1]

# YOUR CODE HERE
````

````python
# SOLUTION — only < instead of >
def insertionSort_desc(a):
    for j in range(1, len(a)):
        key = a[j]
        i = j
        while i > 0 and a[i-1] < key:   # < flips sort order
            a[i] = a[i-1]
            i = i - 1
        a[i] = key
    return a

print(insertionSort_desc([5, 2, 8, 1, 9]))  # [9, 8, 5, 2, 1]
````

# DAY 4 — Stacks & Singly Linked Lists
### Stack = last in, first out (LIFO). Add/remove from the SAME end.
### Singly Linked List (dict style): `llist['head'] -> node -> ... -> 'tail'`

````python
# WORKED EXAMPLE: stack (your exam pattern)
def create(): return []
def push(s, e): s.append(e); return s
def pop_top(s): return s.pop()         # removes AND returns top
def top(s): return s[-1]               # just peeks
def is_empty(s): return len(s) == 0

s = create()
push(s, 10); push(s, 20); push(s, 30)
print(s)          # [10, 20, 30]
print(top(s))     # 30
pop_top(s)
print(s)          # [10, 20]
````

````python
# PRACTICE 11: Use a stack to REVERSE a list. Push all elements, then pop them into a new list.
# reverse_with_stack([1,2,3,4,5]) => [5,4,3,2,1]

# YOUR CODE HERE
````

````python
# SOLUTION
def reverse_with_stack(lst):
    stack = []
    for item in lst:
        stack.append(item)    # push
    result = []
    while stack:
        result.append(stack.pop())  # pop from top
    return result

print(reverse_with_stack([1, 2, 3, 4, 5]))  # [5, 4, 3, 2, 1]
````

````python
# WORKED EXAMPLE: singly linked list with dict (your exam pattern)
# Structure: llist['head'] points to first node, each node points to next, last node points to 'tail'
def create_sll():
    llist = dict()
    llist['head'] = 'tail'
    llist['tail'] = 0
    return llist

def add_front(llist, node):   # adds to front (like a stack — LIFO)
    llist[node] = llist['head']
    llist['head'] = node
    return llist

def traverse(llist):
    current = llist['head']
    while current != 'tail':
        print(current)
        current = llist[current]

l = create_sll()
add_front(l, 10)
add_front(l, 20)
add_front(l, 30)
traverse(l)  # prints 30, 20, 10  (most recent first)
````

````python
# PRACTICE 12: Implement sum_sll(llist) that returns the sum of all numeric nodes.
# Use the same dict structure from above.
# For llist with nodes 10,20,30: sum = 60

# YOUR CODE HERE
````

````python
# SOLUTION
def sum_sll(llist):
    current = llist['head']
    total = 0
    while current != 'tail':
        total += current
        current = llist[current]
    return total

l = create_sll()
add_front(l, 10); add_front(l, 20); add_front(l, 30)
print(sum_sll(l))  # 60
````

# DAY 5 — Doubly Linked Lists (DLL)
### The exam LOVES this. Know it cold.
```
dll = {
  'head': {'prev': None, 'next': 12},
  12: {'prev': 'head', 'next': 20},
  20: {'prev': 12,     'next': 'tail'},
  'tail': {'prev': 20, 'next': None}
}
```
**To insert node X after node A:**
1. B = dll[A]['next']
2. dll[X] = {'prev': A, 'next': B}
3. dll[A]['next'] = X
4. dll[B]['prev'] = X

**To delete node X:**
1. L = dll[X]['prev'],  R = dll[X]['next']
2. dll[L]['next'] = R
3. dll[R]['prev'] = L
4. del dll[X]

````python
# WORKED EXAMPLE: create DLL and insert elements as a QUEUE (enqueue at tail side)
def create_dll():
    return {'head': {'prev': None, 'next': 'tail'},
            'tail': {'prev': 'head', 'next': None}}

def enqueue(dll, x):   # insert before tail
    prev = dll['tail']['prev']
    dll[x] = {'prev': prev, 'next': 'tail'}
    dll[prev]['next'] = x
    dll['tail']['prev'] = x
    return dll

def dequeue(dll):      # remove after head
    first = dll['head']['next']
    if first == 'tail':
        return dll     # empty
    right = dll[first]['next']
    dll['head']['next'] = right
    dll[right]['prev'] = 'head'
    del dll[first]
    return dll

q = create_dll()
enqueue(q, 5); enqueue(q, 10); enqueue(q, 15)
print(q)
dequeue(q)      # removes 5 (first in)
print(q)
````

````python
# PRACTICE 13 (exam-style — like popfirst2 from your notes):
# Implement remove_last2(dll) that removes the TWO nodes immediately BEFORE the tail.
# (mirror of popfirst2 but from the tail side)

listx = {
    'head': {'prev': None, 'next': 10},
    10: {'prev': 'head', 'next': 20},
    20: {'prev': 10, 'next': 30},
    30: {'prev': 20, 'next': 40},
    40: {'prev': 30, 'next': 'tail'},
    'tail': {'prev': 40, 'next': None}
}
# Expected: removes 30 and 40, tail.prev becomes 20

# YOUR CODE HERE
````

````python
# SOLUTION
def remove_last2(dll):
    n1 = dll['tail']['prev']          # node just before tail (40)
    n2 = dll[n1]['prev']              # node before that (30)
    left = dll[n2]['prev']            # node that stays as new last (20)
    dll[left]['next'] = 'tail'
    dll['tail']['prev'] = left
    del dll[n1]
    del dll[n2]
    return dll

listx = {
    'head': {'prev': None, 'next': 10},
    10: {'prev': 'head', 'next': 20},
    20: {'prev': 10, 'next': 30},
    30: {'prev': 20, 'next': 40},
    40: {'prev': 30, 'next': 'tail'},
    'tail': {'prev': 40, 'next': None}
}
print(remove_last2(listx))
# head->10->20->tail
````

````python
# PRACTICE 14 (exam-style — like dllc): Implement dll_max(dll) that returns the MAXIMUM numeric value
# in a doubly linked list. Traverse using references.

dll = {
    'head': {'prev': None, 'next': 7},
    7:  {'prev': 'head', 'next': 42},
    42: {'prev': 7,      'next': 3},
    3:  {'prev': 42,     'next': 19},
    19: {'prev': 3,      'next': 'tail'},
    'tail': {'prev': 19, 'next': None}
}
# Expected: 42

# YOUR CODE HERE
````

````python
# SOLUTION
def dll_max(dll):
    current = dll['head']['next']
    max_val = current                   # first real node is initial max
    while current != 'tail':
        if current > max_val:
            max_val = current
        current = dll[current]['next']
    return max_val

dll = {
    'head': {'prev': None, 'next': 7},
    7:  {'prev': 'head', 'next': 42},
    42: {'prev': 7,      'next': 3},
    3:  {'prev': 42,     'next': 19},
    19: {'prev': 3,      'next': 'tail'},
    'tail': {'prev': 19, 'next': None}
}
print(dll_max(dll))  # 42
````

````python
# PRACTICE 15 (exam-style — like ds7 / stack3): 
# Implement ds_queue(listx, v1) that:
# 1) creates empty DLL
# 2) inserts elements of listx as a QUEUE (FIFO: new at tail)
# 3) adds v1 to the queue
# 4) removes ONE element from the front
# 5) returns the DLL

# YOUR CODE HERE
````

````python
# SOLUTION
def ds_queue(listx, v1):
    dll = {'head': {'prev': None, 'next': 'tail'},
           'tail': {'prev': 'head', 'next': None}}
    listx.append(v1)
    for x in listx:                    # enqueue each element (at tail)
        prev = dll['tail']['prev']
        dll[x] = {'prev': prev, 'next': 'tail'}
        dll[prev]['next'] = x
        dll['tail']['prev'] = x
    # dequeue one element (from head)
    first = dll['head']['next']
    right = dll[first]['next']
    dll['head']['next'] = right
    dll[right]['prev'] = 'head'
    del dll[first]
    return dll

result = ds_queue([5, 10, 15], 20)
for k, v in result.items():
    print(f"{k}: {v}")    # 5 removed; remaining: 10->15->20
````

# DAY 6 — Binary Trees & Graphs
### Binary Tree (dict style):
```
btree = {
  'root': {'P': None,   'L': 1,    'R': 10},
  1:      {'P': 'root', 'L': 5,    'R': 3},
  10:     {'P': 'root', 'L': 12,   'R': 13},
  ...leaves have 'L': None, 'R': None
}
```
**BFS traversal pattern** (used in higher/lower/btc):
```python
queue = [root]
while queue:
    current = queue.pop(0)
    lc = tree[current]['L']
    rc = tree[current]['R']
    if lc: queue.append(lc)
    if rc: queue.append(rc)
    # process current here
```
### Graph (edge list): `[(1,4),(1,3),(2,3)]`
- Count connections: `degsum[u] = degsum.get(u,0) + 1`

````python
# WORKED EXAMPLE: btc — sum all numeric node values (your exam pattern)
btree = {
    'root': {'P': None,   'L': 1,    'R': 10},
    1:      {'P': 'root', 'L': 5,    'R': 3},
    10:     {'P': 'root', 'L': 12,   'R': 13},
    5:      {'P': 1,      'L': None, 'R': None},
    3:      {'P': 1,      'L': None, 'R': None},
    12:     {'P': 10,     'L': None, 'R': None},
    13:     {'P': 10,     'L': None, 'R': None}
}

def btc(tree, root):
    queue = [root]
    total = 0
    while queue:
        current = queue.pop(0)
        lc = tree[current]['L']
        rc = tree[current]['R']
        if lc: queue.append(lc)
        if rc: queue.append(rc)
        if isinstance(current, (int, float)):
            total += current
    return total

print(btc(btree, 'root'))  # 1+10+5+3+12+13 = 44
````

````python
# PRACTICE 16 (exam-style — like higher/lower): Implement between(tree, root, lo, hi)
# that returns all nodes with values BETWEEN lo and hi (inclusive).
# between(btree, 'root', 3, 12) => [10, 5, 3, 12]  (order depends on BFS)

# YOUR CODE HERE
````

````python
# SOLUTION
def between(tree, root, lo, hi):
    queue = [root]
    result = []
    while queue:
        current = queue.pop(0)
        lc = tree[current]['L']
        rc = tree[current]['R']
        if lc: queue.append(lc)
        if rc: queue.append(rc)
        if isinstance(current, (int, float)):
            if lo <= current <= hi:
                result.append(current)
    return result

btree = {
    'root': {'P': None,   'L': 1,    'R': 10},
    1:      {'P': 'root', 'L': 5,    'R': 3},
    10:     {'P': 'root', 'L': 12,   'R': 13},
    5:      {'P': 1,      'L': None, 'R': None},
    3:      {'P': 1,      'L': None, 'R': None},
    12:     {'P': 10,     'L': None, 'R': None},
    13:     {'P': 10,     'L': None, 'R': None}
}
print(between(btree, 'root', 3, 12))  # nodes between 3 and 12
````

````python
# WORKED EXAMPLE: degrees — count connections for each node (your exam pattern)
def degrees(graph):
    degsum = dict()
    for u, v in graph:
        degsum[u] = degsum.get(u, 0) + 1
        degsum[v] = degsum.get(v, 0) + 1
    return degsum

graph = [(1,4),(1,3),(2,3),(2,5)]
print(degrees(graph))  # {1:2, 4:1, 3:2, 2:2, 5:1}
````

````python
# PRACTICE 17 (exam-style — like neighbors2): Implement isolated_nodes(graph) that returns
# nodes that appear in the edge list but have EXACTLY ONE connection.
# isolated_nodes([(1,4),(1,3),(2,3),(2,5)]) => [4, 5]  (each connected to only 1 other node)

# YOUR CODE HERE
````

````python
# SOLUTION
def isolated_nodes(graph):
    degsum = dict()
    for u, v in graph:
        degsum[u] = degsum.get(u, 0) + 1
        degsum[v] = degsum.get(v, 0) + 1
    result = []
    for k in degsum:
        if degsum[k] == 1:
            result.append(k)
    return result

print(isolated_nodes([(1,4),(1,3),(2,3),(2,5)]))  # [4, 5]
````

````python
# SOLUTION Q1
def ds_minlength(xlist):
    result = []
    for inner in xlist:
        if len(inner) > 2:
            min_val = inner[0]
            for num in inner:
                if num < min_val:
                    min_val = num
            result = result + [min_val]
    return result

print(ds_minlength([[5,3],[1,8,2],[4],[7,6,9,0]]))  # [2, 0]
````

````python
# EXAM Q3 (DLL — like stack3): Implement ds_stack(listx, v1) that:
# 1) creates empty DLL
# 2) inserts elements of listx as a STACK (LIFO: new at head)
# 3) adds v1 to the stack
# 4) removes TWO elements from the stack
# 5) returns the DLL

# YOUR CODE HERE
````

````python
# SOLUTION Q3
def ds_stack(listx, v1):
    dll = {'head': {'prev': None, 'next': 'tail'},
           'tail': {'prev': 'head', 'next': None}}
    listx.append(v1)
    for x in listx:              # push: insert after head
        succ = dll['head']['next']
        dll[x] = {'prev': 'head', 'next': succ}
        dll['head']['next'] = x
        dll[succ]['prev'] = x
    for _ in range(2):           # pop twice from head
        top = dll['head']['next']
        right = dll[top]['next']
        dll['head']['next'] = right
        dll[right]['prev'] = 'head'
        del dll[top]
    return dll

result = ds_stack([1, 2, 3], 4)
for k, v in result.items():
    print(f"{k}: {v}")
# Stack was [4,3,2,1] (4 on top), removes 4 and 3, leaves [2,1]
````

````python
# EXAM Q4 (Binary Tree): Implement count_leaves(tree, root) that returns the NUMBER OF LEAF NODES
# (nodes where both L and R are None). Use BFS traversal.

btree = {
    'root': {'P': None,   'L': 1,    'R': 10},
    1:      {'P': 'root', 'L': 5,    'R': 3},
    10:     {'P': 'root', 'L': 12,   'R': 13},
    5:      {'P': 1,      'L': None, 'R': None},
    3:      {'P': 1,      'L': None, 'R': None},
    12:     {'P': 10,     'L': None, 'R': None},
    13:     {'P': 10,     'L': None, 'R': None}
}
# Expected: 4 leaves (5, 3, 12, 13)

# YOUR CODE HERE
````

````python
# SOLUTION Q4
def count_leaves(tree, root):
    queue = [root]
    count = 0
    while queue:
        current = queue.pop(0)
        lc = tree[current]['L']
        rc = tree[current]['R']
        if lc: queue.append(lc)
        if rc: queue.append(rc)
        if lc is None and rc is None:   # it's a leaf
            count += 1
    return count

print(count_leaves(btree, 'root'))  # 4
````

````python
# EXAM Q5 (Graph): Implement hub_nodes(graph) that returns nodes connected to MORE THAN 2 others.
# hub_nodes([(1,2),(1,3),(1,4),(2,3),(4,5)]) => [1, 3]  ...wait, check by hand!
# Degrees: 1:3, 2:2, 3:2, 4:2, 5:1 => only node 1 has > 2

# YOUR CODE HERE
````

````python
# SOLUTION Q5
def hub_nodes(graph):
    degsum = dict()
    for u, v in graph:
        degsum[u] = degsum.get(u, 0) + 1
        degsum[v] = degsum.get(v, 0) + 1
    result = []
    for k in degsum:
        if degsum[k] > 2:
            result.append(k)
    return result

print(hub_nodes([(1,2),(1,3),(1,4),(2,3),(4,5)]))  # [1]
````

````python
# EXAM Q6 (DLL traversal — like great from your notes):
# Implement dll_sorted(dll) that returns True if the nodes are in INCREASING order
# when traversed from head to tail.

dll1 = {
    'head': {'prev': None, 'next': 5},
    5:  {'prev': 'head', 'next': 10},
    10: {'prev': 5,      'next': 20},
    20: {'prev': 10,     'next': 'tail'},
    'tail': {'prev': 20, 'next': None}
}  # True

dll2 = {
    'head': {'prev': None, 'next': 5},
    5:  {'prev': 'head', 'next': 3},
    3:  {'prev': 5,      'next': 10},
    10: {'prev': 3,      'next': 'tail'},
    'tail': {'prev': 10, 'next': None}
}  # False (5 > 3)

# YOUR CODE HERE
````

````python
# SOLUTION Q6
def dll_sorted(dll):
    current = dll['head']['next']
    while dll[current]['next'] != 'tail':
        nxt = dll[current]['next']
        if current > nxt:        # out of order
            return False
        current = nxt
    return True

print(dll_sorted(dll1))  # True
print(dll_sorted(dll2))  # False
````

````python
# EXAM Q7: Implement is_rotation(s1, s2) that returns True if s2 is a rotation of s1.
# A rotation: 'abcde' rotated by 2 = 'cdeab'
# Trick: s2 is a rotation of s1 if s2 is found inside s1+s1 (and same length)
# is_rotation('abcde','cdeab') => True
# is_rotation('abcde','abced') => False

# YOUR CODE HERE
````

````python
# SOLUTION Q7
def is_rotation(s1, s2):
    if len(s1) != len(s2):
        return False
    doubled = s1 + s1
    # check if s2 appears in doubled (without using 'in' for strings is allowed in exam)
    n = len(s2)
    for i in range(len(doubled) - n + 1):
        if doubled[i:i+n] == s2:
            return True
    return False

print(is_rotation('abcde', 'cdeab'))  # True
print(is_rotation('abcde', 'abced'))  # False
````

````python
# SOLUTION Q8
def scalar_product(v1, v2):
    result = []
    if len(v1) != len(v2):
        return 'lengths not equal'
    for i in range(len(v1)):
        result.insert(i, v1[i] * v2[i])
    return result

print(scalar_product([1,2,3],[4,5,6]))  # [4, 10, 18]
````

````python
# EXAM Q9 (DLL — like remfirlas): Implement swap_ends(dll) that swaps the values of
# the first node (after head) and the last node (before tail) WITHOUT creating new nodes.
# Hint: just swap the keys in the structure — actually easier to swap the node VALUES
# by rebuilding the dict. BUT the exam says use references, so swap by relinking.
# Simpler approach: just note the two node keys, detach both, reinsert in swapped positions.

dll = {
    'head': {'prev': None, 'next': 10},
    10: {'prev': 'head', 'next': 20},
    20: {'prev': 10, 'next': 30},
    30: {'prev': 20, 'next': 'tail'},
    'tail': {'prev': 30, 'next': None}
}
# After swap: head->30->20->10->tail

# YOUR CODE HERE
````

````python
# SOLUTION Q9 — relinking approach
def swap_ends(dll):
    first = dll['head']['next']
    last = dll['tail']['prev']
    if first == last:          # only one node, nothing to swap
        return dll
    
    # neighbours of first and last
    first_right = dll[first]['next']
    last_left   = dll[last]['prev']
    
    if first_right == last:    # adjacent nodes — special case
        dll['head']['next'] = last
        dll[last]['prev'] = 'head'
        dll[last]['next'] = first
        dll[first]['prev'] = last
        dll[first]['next'] = 'tail'
        dll['tail']['prev'] = first
    else:
        # disconnect first
        dll['head']['next'] = first_right
        dll[first_right]['prev'] = 'head'
        # disconnect last
        dll['tail']['prev'] = last_left
        dll[last_left]['next'] = 'tail'
        # insert last at head position
        new_first_right = dll['head']['next']
        dll[last] = {'prev': 'head', 'next': new_first_right}
        dll['head']['next'] = last
        dll[new_first_right]['prev'] = last
        # insert first at tail position
        new_last_left = dll['tail']['prev']
        dll[first] = {'prev': new_last_left, 'next': 'tail'}
        dll[new_last_left]['next'] = first
        dll['tail']['prev'] = first
    return dll

dll = {
    'head': {'prev': None, 'next': 10},
    10: {'prev': 'head', 'next': 20},
    20: {'prev': 10, 'next': 30},
    30: {'prev': 20, 'next': 'tail'},
    'tail': {'prev': 30, 'next': None}
}
result = swap_ends(dll)
for k, v in result.items():
    print(f"{k}: {v}")   # head->30->20->10->tail
````

````python
# EXAM Q10 (Mixed — dict + list): A phone book is stored as:
# pb = {'Alice': ['123', 'work'], 'Bob': ['456', 'home'], 'Carol': ['789', 'work']}
# Implement by_type(pb, contact_type) that returns all names with the given contact type.
# by_type(pb, 'work') => ['Alice', 'Carol']

pb = {'Alice': ['123', 'work'], 'Bob': ['456', 'home'], 'Carol': ['789', 'work'], 'Dave': ['000', 'home']}

# YOUR CODE HERE
````

````python
# SOLUTION Q10
def by_type(pb, contact_type):
    result = []
    for name in pb:
        if pb[name][1] == contact_type:
            result += [name]
    return result

pb = {'Alice':['123','work'],'Bob':['456','home'],'Carol':['789','work'],'Dave':['000','home']}
print(by_type(pb, 'work'))   # ['Alice', 'Carol']
print(by_type(pb, 'home'))   # ['Bob', 'Dave']
````
